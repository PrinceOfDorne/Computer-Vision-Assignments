```
import numpy
```

```
#Self Created function for Histogram

def get_histogram(image, bins):

    histogram = numpy.zeros(bins)
  
    for pixel in image:
        histogram[pixel] += 1               #Digital Processing Assignment by Vedic Mishra(2017UEE0089)
  
    return histogram
```

```
#Cummulative Summation

def cumsum(a):
    a = iter(a)
    b = [next(a)]
    for i in a:
        b.append(b[-1] + i)
    return numpy.array(b)
```

```
def Histogram_Equalization(img):

  flat = img.flatten()
  histogram = get_histogram(flat, 256)

  cs = cumsum(histogram)
  nj = (cs-cs.min()) * 255
  N = cs.max() - cs.min()
  cs = nj / N
  cs = cs.astype('uint8')       #Digital Processing Assignment by Vedic Mishra(2017UEE0089)
  img_new = cs[flat]

  bins = 50
  plt.hist(flat, bins, alpha=0.5, label='Original')
  plt.hist(img_new, bins, alpha=0.5, label='New')
  plt.legend(loc='upper right')
  plt.show()

  img_new = numpy.reshape(img_new, img.shape)

  return img_new
```

```
boat_new = Histogram_Equalization(boat)
```

```
dive_new = Histogram_Equalization(dive)
```

```
couple_new = Histogram_Equalization(couple)
```

```
cow_new = Histogram_Equalization(cow)
```

```
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(boat, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(boat_new, cmap='gray')

plt.show(block=True)
```
